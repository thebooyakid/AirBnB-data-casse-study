```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('./AB_NYC_2019.csv')
air_bnb.head(10)
```

```python
# How many neighborhood groups are available and which shows up the most?
neighbourhood_groups = air_bnb.groupby(['neighbourhood_group']).size()
neighbourhood_groups

# 5 neighborhhod groups and Manhattan shows up the most
```

```python
# Are private rooms the most popular in manhattan?
room_type = air_bnb.groupby(['neighbourhood_group','room_type']).size()
room_type

# Entire home/apt are the most popular in Manhattan
```

```python
# Which hosts are the busiest and based on their reviews?

host_id_group = air_bnb.groupby(['host_id','host_name']).sum().sort_values('number_of_reviews',ascending=False)
host_id_group.head(10)

# Maya, host_id 37312959 was the busiest with 2273 reviews
```

```python
#Which neighorhood group has the highest average price?

max_avg_neighbourhood_group = air_bnb.groupby(['neighbourhood_group']).mean().round(decimals=2).sort_values('price',ascending=False) 
max_avg_neighbourhood_group

# Manhattan has the hightst average price at $196.88
```

```python
# Which neighbor hood group has the highest total price?
max_price_neighbourhood_group = air_bnb.groupby(['neighbourhood_group']).sum().sort_values('price',ascending=False) 
max_price_neighbourhood_group

#Manhattan has the highest total price at $4,264,527
```

```python
#Which top 5 hosts have the highest total price?
expensive_host = air_bnb.groupby(['host_id','host_name']).sum()[['price']].sort_values('price',ascending=False)
expensive_host.head(5)

#Sonder (NYC), Blueground, Sally, Red Awning, and Kara have the total highest price
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

# air_bnb[air_bnb['number_of_reviews']>100]

no_avail = air_bnb.sort_values(['availability_365','number_of_reviews']).where(air_bnb['number_of_reviews']>100)
no_avail

# There are a lot of them
```

```python
# What host has the highest total of prices and where are they located?
pricy_host = air_bnb.groupby(['host_id','host_name','neighbourhood_group']).sum()[['price']].sort_values('price',ascending=False)
pricy_host.head(1)

#Sonder(NYC) has the highest total of prices, and is located in Manhattan
```

```python
# When did Danielle from Queens last receive a review?
# latest_review = air_bnb.groupby(['host_name','neighbourhood_group']).sum().sort_values('last_review',ascending=False)
# latest_review
latest_review = air_bnb['last_review'].where(air_bnb['host_name'] == 'Danielle').where(air_bnb['neighbourhood_group'] == 'Queens')
latest_review
```

## Further Questions

1. Which host has the most listings?

```python
most_listings = air_bnb.groupby(['host_id','host_name']).sum().sort_values('id',ascending=False)
most_listings.head(10)
```

2. How many listings have completely open availability?

```python
open_avail = air_bnb.groupby(['id']).mean().where(air_bnb['availability_365'] == 0)
open_avail
```

3. What room_types have the highest review numbers?

```python
high_reviews = air_bnb.groupby(['room_type']).sum().sort_values('number_of_reviews',ascending=False)
high_reviews
```
